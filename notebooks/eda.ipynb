```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
import yfinance as yf
import os

# ✅ Yeh line folder banayega agar nahi hai to
os.makedirs("data", exist_ok=True)

# ✅ Data download karo
data = yf.download('AAPL', start='2005-01-01', end='2025-05-31')

# ✅ Save karo CSV file
data.to_csv('data/aapl.csv')

print("✅ Data saved to data/aapl.csv successfully!")
```
